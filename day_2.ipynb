### 1. Arithmetic Operators 
Arithmetic operators are symbols used in mathematics and programming to perform basic mathematical calculations like addition, subtraction, multiplication, and division. 

```python
a = 9
b = 2

print("The additon of a and b =",a + b)  # Output: 11
print("The subtraction of a and b=",a-b)  # Output: 7
print("The multiplication of a and b=",a * b)  # Output: 18
print("The division of a and b=",a / b)  # Output: 4.5
print("The remainder a/b=",a % b)  # Output: 0 reminder 
#power 
print("When b is taken as the power of a=",a ** b)
#floor division
print("The integer part of division of a and b=",a // b)  # Output: 4
```

### 2. Assingment operators
Assignment operators are symbols used in programming to assign a value to a variable.

     In the program below the new value we assign to y is n addition with its previous value.
     and the value we assign to x is in multiplication with 2

```python
x = 5
y = 10

y += 5     # y = y + 5 
print("The value of y is",y)    # Output: 15

x *= 2      # x = x * 2 
print("The value of x is",x)
```

### 3. Comparison (Relational Operators )
Comparison operators are symbols used to compare two values.
They check whether a condition is true or false.
   
    In the program given below we are comparing two integers on different conditions using      comparison operators.

```python
a = 11
b = 3 

print(" a is equal to b:", a == b )
print(" a is not equal to b:", a != b )
print(" a is greater than b:",a > b )
print(" a is smaller than b:",a < b)
print(" a is greater than or equal to b:", a >= b )
print(" a is smaller than or equal to b:",a <= b)
```

### 4.  Logical Operators
Logical operators are symbols or keywords used to combine or modify conditions.
They return a result of True or False.

   AND is true when both conditions are true .

   OR is true even when one condition is true.
   
   NOT is the reverse of given conditon.

```python


p = 10
q = 4 

print ( "p is greater than 5 and q is smaller than 3:",(p > 5) and (q < 3) )  # True and false  -> false
print ("p is greater than 5 or q is smaller than 3: " ,(p > 5) or (q < 3) )   # True or false  -> true
print ( "p is not greater than 5:",not(p > 5) )            # not true -> false
```

### 5. Membership operatos
Membership operators are used to check whether a value exists in a sequence (such as a list, string, tuple, or set).
In Python, the membership operators are:

in → Returns True if the value is found in the sequence

not in → Returns True if the value is not found in the sequence


```python
# Membership Operators

fruits = [ "banana", "cherry"] #this is a list named fruits.
```

```python
print("apple is present in fruits:","apple"  in fruits)  
print("banana is present in fruits:","banana"in fruits)
```

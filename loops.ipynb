# Loops
A loop in Python is a control flow construct that enables the repeated execution of a block of statements as long as a specified condition holds true or for each element in a given sequence.

### Python provides two primary looping mechanisms:

The for loop, which iterates over elements of a sequence (such as lists, tuples, strings, dictionaries, or ranges).

The while loop, which repeatedly executes a block of code based on a Boolean condition.

```python
for i in range(5):
    print(i)
```

```python
count = 0

while count < 5:
    print(count)
    count += 1
```

### Break statement
Stops the loop completely.

```python
name="kabisha mir"

for i in range(7):
    if i==5:
       break
    print(name)
    
```

```python
name="kabisha mir"
for i in range (7):
    if i==3:
        break
    for letter in name:
        print(letter)
```

```python
string="aabbcc"
count=3
while True:
    if string[0]=='a':
        string=string[2:]
    elif string[-1]=='b':
        string=string[:2]
    else:
        count+=1
        break
    print(string)
    print(count)
```

### Continue statement
Skips the iteration of current loop

```python
i = 0

while i < 6:
    i += 1
    
    if i == 3:
        continue
    
    print(i)
```

```python
name="kabisha"
i=1
while i<10:
    i+=1
    if i==5:
        continue
    print(name)
```

```python
for i in range(1, 6):
    if i == 3:
        continue
    print(i)
```

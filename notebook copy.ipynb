# Variable

#### A variable in python is like a named box in which you can store some data and use it later.

Syntax:

```python
variable_name= value
```

Example:

```python
age=18
```

### Rules of variable naming:

1. Variable name must start with a letter or underscore(_)
2. It cannot start with a number.
3. Variable names cannot contain spaces.

```python
total_marks=90 #correct example
1 total=90    #incorrect example
```

# Integer Data Type

An integer data stores whole numbers(postive or nrgative) without decimals.

#### Example:

```python
num = 25
```

## String Data Type

A string data type is used to store text or characters.

##### Two ways to create strings:

```python
name = "Kabisha"
city = 'Lahore'
```

## Float Data Type

A float data type stores numbers with decimal points.

#### Example:

```python
price = 99.9
```

## Boolean Data Type

A boolean data type stores true or false values.

#### Possible values

1. True
2. False

#### Example

```python
is_student = True
```

# type() Function

A type function is used to check data type of a variable.

#### Example:

```python
x = 10
print(type(x))
```

# input() Function

This function is used to take input from the user.

Its default data type is string(str.)

#### Example:

```python
name = input("Enter your name: ")
```

# Type casting

This means converting one data type to another.

#### Example

```python
num = 10
num = int(num)
```

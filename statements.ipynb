# Match statements 
A match statement in Python is a way to check a value and run different code depending on what it matches.


### Programs using match statements

```python
day=int(input("Enter day:"))
match day:
    case 1:
        print("Monday")
    case 2:
        print("Tuesday")
    case 3:
        print("Wednesday")
    case 4:
        print("Thursday")
    case 5:
        print("Friday")
    case 6:
        print("Saturday")
    case 7:
        print("Sunday")
    case _:
        print("day does not exist.")
        
```

```python
a=10
b=12
operator=input("Enter the operator of your choice(+,-,*,/)")
match operator:
    case "+" :
        print("a+b=",a+b)
    case "-":
        print("a-b=",a-b)
    case "*":
        print("a*b=",a*b)
    case "/":
        print("a/b=",a/b)
    case _:
        print("Invalid operator")
        
```

```python
color=input("Enter color of traffic light:")
match color:
    case "red":
        print ("stop")
    case "green":
        print("go")
    case "yellow":
        print("ready")
    case _:
        print("Not a color of traffic light")

```

## Difference between if else and match statements

### Purpose

if-else is used for evaluating logical conditions.

match is used for pattern or value matching.

### Flexibility

if-else can evaluate complex boolean expressions.

match works best with predefined values or patterns.

### Version Requirement

if-else is available in all Python versions.

match is available only in Python 3.10 and later.

### Use Case

if-else is preferred for condition-based logic (e.g., checking greater than or less than).

match is preferred when handling multiple fixed cases (e.g., menu options, operators, specific values).

In summary, if-else is used for conditional evaluation, while match is used for structured pattern matching.

## Programs using nested if else

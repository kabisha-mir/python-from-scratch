# Variables
Variables are used to store data in a program.They act as containers for values that can be changed or manipulated during the execution of a program.
## Syntax:
variable_name = value
### Example:
age = 18
name = "Ayaan"
* (Here 'age' ia a variable holding a numeric value , and 'name' is a variable holding text)

### Practical code Explanation:
* In the code below, there are two variables.
* We use the = sign (Assignment Operator) to store the value -2 and "john david" into these variables.
* These variables are act as containers that hold our data so we can use it later in the print function.

```python
# i store student name in std_n variable.
std_n =  "john david" 

#storing student roll number in variable 'student_roll_number
student_roll_number = -2 


print(std_n)
print(student_roll_number)  
```

# Datatypes
 #### 1. Integer : 
An Integer (int) is a fundamental data type in python that is used to represent whole numbers.It includes all positive and negative numbers including zero, but it does not contain any decimal or fractional parts.  

```python
# Example of integer 
student_roll_number = 2
student_age = 20
```

#### 2. float :
A float (floating-point number) is a data type in Python used to represent numbers that have decimal point. 

```python
# Example of float
student_gpa = 3.5
```

#### 3. string :
 A string (str) is a data type in python used to represent text. It is a sequence of characters wrapped inside quotes.
**Two ways to create a string :**
* ***Single-line String :***  Created using double quotes (" ") or single quotes(' ').
* ***Multi-line string:*** Created using triple quotes(""" """). This is used when the text spans across multiple lines.

```python
#string 
# Example of single line string 
student_name = "Ali raza"
```

```python
# Example of multiline string 
ali_intro = """kjdlfkjdlkfj
kldjflsdjf
kldjflsdjf
jlkdsjflsdjf
ldjflsdkjf
"""
```
